```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Ticker da Sanepar na B3 é "SAPR11.SA"
ticker = "SAPR11.SA"

# Baixar os dados históricos da Sanepar
data = yf.download(ticker, start="2023-01-01", end="2024-09-01")

# Exibir as primeiras linhas dos dados
print(data.head())

# Análise simples: cálculo da média móvel de 20 dias
data['20_MA'] = data['Close'].rolling(window=20).mean()

# Plotar os preços de fechamento e a média móvel
plt.figure(figsize=(14,7))
plt.plot(data['Close'], label='Preço de Fechamento')
plt.plot(data['20_MA'], label='Média Móvel de 20 dias', color='orange')
plt.title('Análise de Preços de Fechamento da Sanepar com Média Móvel de 20 dias')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.legend()
plt.grid(True)
plt.show()
```
