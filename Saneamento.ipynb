```python

import matplotlib.pyplot as plt

# Função para capturar os dados
def capturar_dados():
    dados = {
        'Volume Produzido': float(input("Informe o Volume Produzido (em m³): ")),
        'Volume Distribuído': float(input("Informe o Volume Distribuído (em m³): ")),
        'Volume Importado': float(input("Informe o Volume Importado (em m³): ")),
        'Volume Exportado': float(input("Informe o Volume Exportado (em m³): ")),
        'Volume Especial': float(input("Informe o Volume Especial (em m³): ")),
        'Perdas': float(input("Informe o Volume de Perdas (em m³): "))
    }
    return dados

# Função para exibir os dados no gráfico
def exibir_grafico(dados):
    categorias = list(dados.keys())
    volumes = list(dados.values())

    plt.figure(figsize=(10,6))
    plt.bar(categorias, volumes, color=['blue', 'green', 'red', 'orange', 'purple', 'brown'])
    plt.title('Volumes da Empresa de Saneamento')
    plt.xlabel('Categorias')
    plt.ylabel('Volume (m³)')
    plt.show()

# Função principal do sistema
def sistema_saneamento():
    print("Sistema de Gestão de Volumes - Empresa de Saneamento")

    # Capturar os dados do usuário
    dados = capturar_dados()

    # Exibir o gráfico
    exibir_grafico(dados)

# Executar o sistema
if __name__ == "__main__":
    sistema_saneamento()
```
